```python
import matplotlib.pyplot as plt
import numpy as np

def plot_feasibility_and_makespan(x_vals, feasibility_data, makespan_data,
                                  title_prefix="STNU-based hybrid approach"):
    """
    Plots feasibility ratio and average makespan side by side for a given approach.

    Parameters:
    - x_vals: list of float (e.g. [1, 0.9, 0.75, 0.5]) – quantiles or x-axis values
    - feasibility_data: list of float – feasibility ratio values
    - makespan_data: list of float – average makespan values
    - title_prefix: str – descriptive prefix for subplot titles
    """
    x_labels = [str(x) for x in x_vals]

    # Set the figure size
    plt.figure(figsize=(12, 6))

    # Feasibility Ratio Plot
    plt.subplot(1, 2, 1)
    plt.plot(x_vals, feasibility_data, marker='o', color='skyblue', linestyle='-', alpha=0.7)
    plt.title(f'Feasibility Ratio of {title_prefix}')
    plt.xlabel('Quantile')
    plt.ylabel('Feasibility Ratio')
    plt.ylim(0, 1)
    plt.xticks(x_vals, x_labels)
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    for x, y in zip(x_vals, feasibility_data):
        plt.text(x, y - 0.05, f"{y:.2f}", ha='center', va='bottom')

    # Average Makespan Plot
    plt.subplot(1, 2, 2)
    plt.plot(x_vals, makespan_data, marker='o', color='salmon', linestyle='-', alpha=0.7)
    plt.title(f'Average Makespan of {title_prefix}')
    plt.xlabel('Quantile')
    plt.ylabel('Average Makespan')
    plt.ylim(0, max(makespan_data) * 1.1)
    plt.xticks(x_vals, x_labels)
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    for x, y in zip(x_vals, makespan_data):
        plt.text(x, y + 0.5, f"{y:.2f}", ha='center', va='bottom')

    plt.tight_layout()
    plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

def plot_grouped_bars(metrics_dict, modes, method_names, colors, title="Comparison Across Modes"):
    """
    Plots grouped bar charts for multiple metrics across different scheduling methods.

    Parameters:
    - metrics_dict: dict[str, list[list[float]]]
        Dictionary where keys are metric names and values are lists of method-specific data.
    - modes: list[str]
        Labels for the x-axis (e.g., ["robust", "quantile_0.9", "quantile_0.75", "quantile_0.5"])
    - method_names: list[str]
        Names of the methods (e.g., ["Proactive", "Reactive", "STNU"])
    - colors: list[str]
        Color for each method (must match length of method_names)
    - title: str
        Title for the full figure
    """
    num_metrics = len(metrics_dict)
    x = np.arange(len(modes))
    bar_width = 0.25

    fig, axs = plt.subplots(2, 2, figsize=(14, 10))
    axs = axs.flatten()

    for i, (metric_name, data_lists) in enumerate(metrics_dict.items()):
        ax = axs[i]
        for j, (method_data, method) in enumerate(zip(data_lists, method_names)):
            ax.bar(x + j * bar_width, method_data, bar_width,
                   label=method, color=colors[j], alpha=0.8)

        ax.set_title(metric_name)
        ax.set_xticks(x + bar_width)
        ax.set_xticklabels(modes, rotation=15)
        ax.set_xlabel("Mode")
        ax.set_ylabel(metric_name)
        ax.grid(axis='y', linestyle='--', alpha=0.5)
        ax.legend()

    plt.suptitle(title, fontsize=16)
    plt.tight_layout(rect=[0, 0.03, 1, 0.95])
    plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

def plot_line_metrics(x_vals, modes, metrics, method_names, colors, markers, title="Line Plot Comparison"):
    """
    Plots line graphs for multiple metrics across methods and quantile modes.

    Parameters:
    - x_vals: list[float] – numeric quantile values (e.g. [1.0, 0.9, 0.75, 0.5])
    - modes: list[str] – labels for the x-axis (e.g. ["robust", "quantile_0.9", ...])
    - metrics: dict[str, list[list[float]]] – keys are metric names, values are lists of method-specific values
    - method_names: list[str] – method labels (e.g. ["Proactive", "Reactive", "STNU"])
    - colors: list[str] – line colors for each method
    - markers: list[str] – marker styles for each method
    - title: str – plot title
    """
    fig, axs = plt.subplots(2, 2, figsize=(14, 10))
    axs = axs.flatten()

    for i, (metric_name, data_lists) in enumerate(metrics.items()):
        ax = axs[i]
        for j, (method_data, method) in enumerate(zip(data_lists, method_names)):
            ax.plot(x_vals, method_data, label=method, marker=markers[j], color=colors[j], linewidth=2)

        ax.set_title(metric_name)
        ax.set_xticks(x_vals)
        ax.set_xticklabels(modes)
        ax.set_xlabel("Mode")
        ax.set_ylabel(metric_name)
        ax.grid(True, linestyle='--', alpha=0.5)
        ax.legend()

    plt.suptitle(title, fontsize=16)
    plt.tight_layout(rect=[0, 0.03, 1, 0.95])
    plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

def compare_datasets_bar_chart(metrics_data, categories, group_labels, title="Comparison Across Settings"):
    """
    Plots a grouped bar chart to compare multiple metrics across different datasets/settings.

    Parameters:
    - metrics_data: dict[str, list[list[float]]]
        A dictionary mapping metric names (e.g., "Makespan") to a list of lists,
        where each inner list corresponds to values for one group (e.g., noise factor 1, noise factor 2).
    - categories: list[str]
        The labels on the x-axis (e.g., ["robust", "quantile_0.9", "quantile_0.75", "quantile_0.5"]).
    - group_labels: list[str]
        Labels for the dataset groups (e.g., ["Noise Factor 1", "Noise Factor 2"]).
    - title: str
        Title for the overall figure.
    """
    num_groups = len(group_labels)
    num_categories = len(categories)
    x = np.arange(num_categories)
    bar_width = 0.8 / num_groups

    colors = plt.cm.tab10.colors  # up to 10 distinguishable colors

    fig, axs = plt.subplots(1, len(metrics_data), figsize=(6 * len(metrics_data), 6))

    if len(metrics_data) == 1:
        axs = [axs]  # make iterable if only one subplot

    for ax, (metric_name, datasets) in zip(axs, metrics_data.items()):
        for i, (data, label) in enumerate(zip(datasets, group_labels)):
            ax.bar(x + i * bar_width, data, bar_width, label=label, color=colors[i % len(colors)], alpha=0.85)

        ax.set_title(metric_name)
        ax.set_xticks(x + bar_width * (num_groups - 1) / 2)
        ax.set_xticklabels(categories, rotation=15)
        ax.set_xlabel("Mode")
        ax.set_ylabel(metric_name)
        ax.grid(axis='y', linestyle='--', alpha=0.5)
        ax.legend()

    plt.suptitle(title, fontsize=16)
    plt.tight_layout(rect=[0, 0.03, 1, 0.95])
    plt.show()

import matplotlib.pyplot as plt
import numpy as np

def compare_datasets_line_plot(metrics_data, x_vals, x_labels, group_labels, title="Line Plot Comparison Across Settings"):
    """
    Plots line graphs to compare multiple datasets across different metrics.

    Parameters:
    - metrics_data: dict[str, list[list[float]]]
        Dictionary where each key is a metric name (e.g., "Avg Makespan") and each value is a list of lists.
        Each inner list represents data for one group (e.g., Noise Factor 1, Noise Factor 2).
    - x_vals: list[float]
        Numeric x-axis values (e.g., [1.0, 0.9, 0.75, 0.5]).
    - x_labels: list[str]
        Labels for the x-axis ticks (e.g., ["robust", "quantile_0.9", "quantile_0.75", "quantile_0.5"]).
    - group_labels: list[str]
        Names of the datasets being compared (e.g., ["Noise Factor 1", "Noise Factor 2"]).
    - title: str
        Title for the entire figure.
    """
    num_groups = len(group_labels)
    colors = plt.cm.tab10.colors
    markers = ['o', 's', '^', 'D', '*', 'x', 'P', 'v']

    fig, axs = plt.subplots(1, len(metrics_data), figsize=(6 * len(metrics_data), 5))

    if len(metrics_data) == 1:
        axs = [axs]  # Ensure axs is iterable

    for ax, (metric_name, data_lists) in zip(axs, metrics_data.items()):
        for i, (data, label) in enumerate(zip(data_lists, group_labels)):
            ax.plot(x_vals, data, label=label,
                    marker=markers[i % len(markers)],
                    color=colors[i % len(colors)],
                    linewidth=2)

        ax.set_title(metric_name)
        ax.set_xticks(x_vals)
        ax.set_xticklabels(x_labels, rotation=15)
        ax.set_xlabel("Mode")
        ax.set_ylabel(metric_name)
        ax.grid(True, linestyle='--', alpha=0.5)
        ax.legend()

    plt.suptitle(title, fontsize=16)
    plt.tight_layout(rect=[0, 0.03, 1, 0.95])
    plt.show()
```
